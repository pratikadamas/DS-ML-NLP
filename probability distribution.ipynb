Binomial Distribution
Binomial Distribution is a Discrete Distribution.

It describes the outcome of binary scenarios, e.g. toss of a coin, it will either be head or tails.

It has three parameters:

n - number of trials.

p - probability of occurence of each trial (e.g. for toss of a coin 0.5 each).

size - The shape of the returned array

```python
# Given 10 trials for coin toss generate 10 data points:



x = random.binomial(n=10, p=0.5, size=10)

print(x)
```

standered normal distribution

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm

# Given x values (symmetric around 0)
x_values = np.array([1, 2, 3, 4, 5, 6, 7, 8, 0, -1, -2, -3, -4, -5, -6, -7, -8])

# Standard normal distribution PDF values
y_values = norm.pdf(x_values, loc=0, scale=1)

# Create a DataFrame for display
df = pd.DataFrame({'x': x_values, 'Standard Normal PDF (y)': y_values})

# Sort for better viewing
df_sorted = df.sort_values(by='x').reset_index(drop=True)

# Plot the values
plt.figure(figsize=(10, 5))
plt.plot(df_sorted['x'], df_sorted['Standard Normal PDF (y)'], marker='o')
plt.title("Standard Normal Distribution Curve")
plt.xlabel("x (Z-score)")
plt.ylabel("Probability Density (y)")
plt.grid(True)
plt.show()

df_sorted
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate normal data
mu, sigma = 0, 1
x = np.random.normal(mu, sigma, 1000)

# Convert to log-normal
y = np.exp(x)

# Convert back to normal
x_back = np.log(y)

# Plot all three
plt.figure(figsize=(12, 6))

# Normal
plt.subplot(1, 3, 1)
sns.histplot(x, kde=True, bins=30, color='skyblue')
plt.title("Normal Distribution")

# Log-normal
plt.subplot(1, 3, 2)
sns.histplot(y, kde=True, bins=30, color='lightgreen')
plt.title("Log-normal Distribution")

# Back to Normal
plt.subplot(1, 3, 3)
sns.histplot(x_back, kde=True, bins=30, color='salmon')
plt.title("Log(y) = Normal again")

plt.tight_layout()
plt.show()
```

central limit theorem

```python


# Population: Exponential (not normal)
population = np.random.exponential(scale=2, size=100000)
plt.hist(population, bins=50, color='red', edgecolor='black', density=True)
plt.title("Sampling Distribution population (CLT)")
plt.xlabel("Population Mean")
plt.ylabel("Density")
plt.grid(True)
plt.show()
print("Population mean ",np.mean(population))

sample_means = []

# Take 1000 samples of size 30
for _ in range(1000):
    sample = np.random.choice(population, size=30)
    sample_means.append(np.mean(sample))

# Plot the result
plt.figure(figsize=(10, 5))
plt.hist(sample_means, bins=50, color='skyblue', edgecolor='black', density=True)
plt.title("Sampling Distribution of Sample Means (CLT)")
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.grid(True)
plt.show()
print("sample mean ",np.mean(sample_means))

```
