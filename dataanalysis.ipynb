numpy vectorized opration 

```python
a1=np.array([1,2,3,4])
a2=np.array([10,20,30,40])

print("Addition elementwise  :: ",a1+a2)
print("substraction elementwise  :: ",a1-a2)
print("Multiplication elementwise  :: ",a1*a2)
print("Divition elementwise  :: ",a1/a2)
```

Statistical concept

```python
data=np.array([1,2,3,4,5,6])
mean=np.mean(data)
std=np.std(data)
normalized_data=(data-mean)/std
print("normalized data is",normalized_data)
print("median data is",np.median(data))
print("Stabderd deiationcdata is",np.std(data))
print(" varience data is",np.var(data))
```

```python
import matplotlib.pylab as plt
```

```python
x=[1,2,3,4,5]
y=[1,4,9,15,25]
#customise lineploat
plt.plot(x,y,color='red',linestyle='--',marker="o")
plt.grid(True)
```

```python
#multiple ploat
x=[1,2,3,4,5]
y=[1,4,9,16,25]
z=[1,2,3,4,5]
plt.figure(figsize=(9,5))
plt.subplot(1,2,1)
plt.plot(x,y)

plt.figure(figsize=(9,5))
plt.subplot(1,2,1)
plt.plot(y,x)
```

```python
#barploat
categoty=['a','s','d','f']
val=[5,3,7,10]
plt.bar(categoty,val, color="green")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("BAR ploat")

```
