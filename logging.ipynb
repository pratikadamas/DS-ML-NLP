Logging in Python is a way to track events that occur while your program is running. It is a useful tool for debugging, troubleshooting, and monitoring your program. The logging module is part of the Python standard library, so you don't need to install anything to use it. 

```python
import logging
# Create and configure logger
logging.basicConfig(filename="newfile.log",
                    format='%(asctime)s-%(name)s-%(levelname)s%-%(message)s',
                    filemode='w')
```
